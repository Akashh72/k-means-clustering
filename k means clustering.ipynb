```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
x, y_true = make_blobs(n_samples=300,centers = 5,cluster_std = 0.50,random_state = 15)
plt.scatter(x[:,0],x[:,1])
```

```python
#centroids delceration
k=5
centroids=[]
n=np.shape(x)[1]
centroids=np.zeros(shape=(k,n),)
```

```python
#centroids assigning
for i in range(k):
    centroids[i]=x[np.random.randint(np.shape(x)[0])]
```

```python
#scatter plotting the centroids
for i in range(k):
    plt.scatter(centroids[i,0],centroids[i,1])
```

```python
for iter in range(1000):
    d={}
    classification=[]    
    
    for i in range(k):
        d[i]=[]
    for data in x:
        distance=[]
        for i in range(k):
            distance.append(np.sum((data-centroids[i])**2)**1/2)
        idx=distance.index(min(distance))
        d[idx].append(data)
        classification.append(idx)
    for i in range(k):
        centroids[i]=np.average(d[i],axis=0)
print(centroids)
```

```python
plt.scatter(x[:,0],x[:,1],c=classification)
plt.scatter(centroids[:,0],centroids[:,1],c='red',s=100)
```
